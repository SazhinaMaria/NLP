```python
s = requests.Session() 

head = ['comments', 'titles','sentiments', 'nicemarks', 'badmarks']

with open('data.csv', mode="w", encoding='utf-8') as f:
     writer  = csv.writer(f)
     writer.writerow(head)

def load_comment_page(page, session):
    
    url = 'https://tutortop.ru/school-reviews/geekbrains/'
    request = session.get(url)    
    return request.text


# Задаем признаки нашего датасета и сохраняем записи в датасет
def contain_comment_data(text):
        
    soup = BeautifulSoup(text, features="lxml")     
    
    comments_blocks = soup.findAll('div', {'class': 'reviews-list-item div_reviews'})
    
    comments=[]
    titles=[]
    sentiments=[]
    nicemarks=[]
    badmarks=[]
    
    it = 0
    for items in comments_blocks:
        
        data = []
        
        comment = items.find('p').text               
        data.append(comment)
        comments.append(comment)
        
        title = items.find('h2').text               
        data.append(title)
        titles.append(title)
        
        sentiment = items.find('span').text               
        data.append(sentiment)
        sentiments.append(sentiment)
        
        nicemark = items.find('span', {'class': 'span_like2'}).text               
        data.append(nicemark)
        nicemarks.append(nicemark)
        
        badmark = items.find('span', {'class': 'span_like'}).text               
        data.append(badmark)
        badmarks.append(badmark)
        
        
        it+=1
        with open('data.csv',  mode="a", encoding='utf-8') as f:
            writer = csv.writer(f)
            writer.writerow(data)
    
    return comments_blocks


page = 0
page_content = load_comment_page(page, s)    
contain_comment_data(page_content)

print("data saved")
```

#### При помощи библиотеки NLTK (Natural Language Toolkit) очистим наш датасет. 

```python
nltk.download('stopwords')
```

```python
df = pd.read_csv('data.csv')
df.info()
```

```python
df.head(5)
```

```python
stop_words = stopwords.words("russian")
stop_words
```

```python
# Выполним очистку для текстов комментариев
df.loc[:,"comments"] = df.comments.apply(lambda x : str.lower(x))
df.loc[:,"comments"] = df.comments.apply(lambda x : " ".join(re.findall('[\w]+',x)))

# Очистим датасет от стоп-слов (cлов, которые не придают смыслового значения предложению). 
# Например: предлоги, союзы, частицы, местоимения и др.
def remove_stopWords(s):
    s = ' '.join(word for word in s.split() if word not in stop_words)
    return s

df.loc[:,"comments"] = df.comments.apply(lambda x: remove_stopWords(x))

def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result

df.loc[:,"comments"] = df.comments.apply(lambda x : remove_numbers(x))

morph = pymorphy2.MorphAnalyzer()

# Выполним лемматизацию (приведем словоформы к их словарной форме - лемме).
# Например: глаголы - к инфинитиву, сущ. и прил. - к им.п., ед.ч.
def lemmatize(s):    
    s = ' '.join(morph.parse(word)[0].normal_form for word in s.split())
    return s

df.loc[:,"comments"] = df.comments.apply(lambda x: lemmatize(x))

# Удалим знаки пунктуации
def remove_punctuation(text):    
    result = re.sub(r'[^\w\s]', '', text)
    return result

df.loc[:,"comments"] = df.comments.apply(lambda x : remove_punctuation(x))

df.to_csv("data_after_preprocess_jupiter.csv")
df['comments'][:10]
```

```python
# Выполним очистку для текстов заголовков
df.loc[:,"titles"] = df.titles.apply(lambda x : str.lower(x))
df.loc[:,"titles"] = df.titles.apply(lambda x : " ".join(re.findall('[\w]+',x)))

def remove_stopWords(s):
    s = ' '.join(word for word in s.split() if word not in stop_words)
    return s

df.loc[:,"titles"] = df.titles.apply(lambda x: remove_stopWords(x))

def remove_numbers(text):
    result = re.sub(r'\d+', '', text)
    return result

df.loc[:,"titles"] = df.titles.apply(lambda x : remove_numbers(x))

morph = pymorphy2.MorphAnalyzer()

def lemmatize(s):    
    s = ' '.join(morph.parse(word)[0].normal_form for word in s.split())
    return s

df.loc[:,"titles"] = df.titles.apply(lambda x: lemmatize(x))

def remove_punctuation(text):    
    result = re.sub(r'[^\w\s]', '', text)
    return result

df.loc[:,"titles"] = df.comments.apply(lambda x : remove_punctuation(x))

df.to_csv("data_after_preprocess_jupiter.csv")
df['titles'][:10]
```
